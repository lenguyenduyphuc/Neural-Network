```python
b = {}
for w in words:
  ch = ['.'] + list(w) + ['.']
  for w in range(len(ch) - 2):
    ch1 =  ch[w] + ch[w+1]
    ch2 = ch[w+2]
    trigram = (ch1, ch2)
    b[ch1, ch2] = b.get(trigram, 0) + 1

sorted(b.items(), key= lambda kv: -kv[1])
```

```python
a = []
c = []
for (first, second), value in b.items():
  a.append(first)
  c.append(second)
```

```python
uni=sorted(set(c))
k=[]
for i in range(len(uni)):
  for j in range(1,len(uni)):
    k.append(''.join(uni[i]+uni[j]))
len(k)
```

```python
stoi_1 = {s:i for i,s in enumerate(uni)}
stoi_1
stoi_2 = {s:i for i,s in enumerate(k)}
stoi_2

itos_1 = {i:s for s,i in stoi_1.items()}
itos_2 = {i:s for s,i in stoi_2.items()}
itos_2
```

```python
for i in range(10):
  ix = 0
  px = 0
  out = []
  while True:
    p = N[px].float()
    p = p/ p.sum()
    ix = torch.multinomial(p, num_samples = 1, replacement=True, generator=g).item()
    out.append(itos_1[ix])
    if ix == 0:
      break
    px = stoi_2[''.join(itos_2[px][1]+itos_1[ix])]

  print(''.join(out))
```

```python
P=N.float()
P /= P.sum(1,keepdim=True)

P[0].sum()
```

```python
loglikelihood=0.0
nll=0.0

n=0
for w in words:
  chs=['.']+list(w)+['.']
  for w in range(len(ch)-2):
    ch1=ch[w]+ch[w+1]
    ch2=ch[w+2]
    ix1=stoi_2[ch1]
    ix2=stoi_1[ch2]
    prob=P[ix1,ix2]
    logprob=torch.log(prob)
    loglikelihood+=logprob
    n+=1
    #print(f'ch1,ch2, logprob={logprob : .4f}')
print(f'loglikelihood={loglikelihood}')
nll=-loglikelihood #negative loglikelihood
print(f'nll={nll}')
nll=nll/n #average negative loglikelihood: loss function
print(f'nll={nll}')
```
